# AVARAGE HOUSE PRICE (INFLATION ADJUSTED),
# AVARAGE INCOME (INFLATION ADJUSTED),
# INFLATION,
# CPI MULTIPLIER,

```python
import pandas as pd
```

```python
df_inflation = pd.read_csv('united-kingdom-inflation-rate-cpi.csv',skiprows=16)
df_inflation = df_inflation[['date',' Inflation Rate (%)']]
df_inflation=df_inflation.rename(columns={' Inflation Rate (%)':'Inflation'})
df_inflation['date']= df_inflation['date'].str.replace('\-.+','',regex=True)

df_inflation.head(20)
```

```python
df_cpi = pd.read_csv('GBRCPIALLMINMEI.csv')
df_cpi['DATE'] = df_cpi['DATE'].str.replace('\-.+','',regex=True)
df_cpi = df_cpi.rename(columns={'DATE':'date'})
df_cpi = df_cpi.rename(columns={'GBRCPIALLMINMEI':'CPI'})
df_cpi = df_cpi.groupby('date').mean().reset_index()
df_cpi['CPI_Multiplier'] = df_cpi['CPI'].iloc[-1] / df_cpi['CPI']

df_cpi.head(20)
```

```python
df_income = pd.read_csv('AWEPPUKQ.csv')
df_income['DATE'] = df_income['DATE'].str.replace('\-.+','',regex=True)
df_income = df_income.groupby('DATE').mean().reset_index()
df_income = df_income.rename(columns={'AWEPPUKQ':'Income'})
df_income = df_income.rename(columns={'DATE':'date'})
df_income['Income'] = df_income['Income'] * 52
df_income = pd.merge(df_income, df_cpi, how='left', on='date')
df_income['CPIAdjIncome'] = df_income['Income'] * df_income['CPI_Multiplier']
df_income = df_income.dropna()
df_income = df_income[['date','Income','CPIAdjIncome']]

df_income
```

```python
df_house_prices = pd.read_csv('Average-prices-2020-10.csv')
df_house_prices = df_house_prices[df_house_prices.values  == "United Kingdom"]
df_house_prices = df_house_prices[['Date','Average_Price']]
df_house_prices['Date'] = df_house_prices['Date'].str.replace('\-.+','',regex=True)
df_house_prices = df_house_prices.groupby('Date').mean().reset_index()
df_house_prices = df_house_prices.rename(columns={'Date':'date'})
df_house_prices = pd.merge(df_house_prices, df_cpi, how='left', on='date')
df_house_prices['CPIAdjPrice'] = df_house_prices['Average_Price'] * df_house_prices['CPI_Multiplier']
df_house_prices = df_house_prices[['date','Average_Price','CPIAdjPrice']]

df_house_prices
```

```python
df_housing = df_house_prices.merge(df_income, on='date').merge(df_inflation, on='date')
df_housing
```

```python
df_housing.to_csv('UK_housing_vs_income.csv')
```
